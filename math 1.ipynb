Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1

```python
x = np.arange(-25, 25,0.1)
y = np.cos(x)
y2 = np.cos(-0.25*x)
```

```python
plt.plot(x, y, label="cos(x)")
plt.plot(x, y2, label="cos(-0.25*x)")

plt.title("Графики функций")
plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")
plt.axis([-25, 25, -1, 1])

plt.show()
```
